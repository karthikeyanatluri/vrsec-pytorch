```python
class Model(nn.Module):
  #input layers
  def __init__(self,in_features=4,h1=8,h2=9,out_features=3):
    super().__init__()
    self.fc1 = nn.Linear(in_features,h1)#fully connected layer 1
    self.fc2 = nn.Linear(h1,h2)#layer 2 connected
    self.out = nn.Linear(h2,out_features)#output layer
    self.fc1 = nn.Linear(in_features,h1)
    self.fc2 = nn.Linear(h1,h2)
    self.out = nn.Linear(h2,out_features)
    super().__init__()

    def forward(self,x):
      x = F.relu(self.fc1(x))#activation function for layer 1
      x = F.relu(self.fc2(x))#for layer 2
      x = self.out(x)#output layer
      return x
```
